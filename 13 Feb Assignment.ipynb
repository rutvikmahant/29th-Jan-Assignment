```python
#Que2
import inspect as ipt  
def tree(self, a = 0):  
    print ('_' * a, self.__name__)  
    for i in self.__subclasses__():  
        tree(i, a + 3)  
print ("The Hierarchy for inbuilt exceptions is: ")    
ipt.getclasstree(ipt.getmro(Exception))  
tree(Exception)
```

```python
#Que4
""" LookupError class is used when a value or index is not found. It has primarily two subclasses KeyError and IndexError. """
# KeyError Example- Here the code is written to ask for a key from user so that it can match it in the dictonary and give output. If not matched KeyError occurs.
student = {'Name':'Manan', 'Age': 28, 'Language': 'Python'}
a = input("Information Required:")

try:
    print(a,"is",student[a])
except LookupError as e:
    print("Error:",e.__class__)

# IndexError Example- Here index given is much higher than number of elemwwents in the given list hence IndexErros is thrown.
lisst = [1,2,3,4,5,6]
try:
    print(lisst[11])
except LookupError as e:
    print("Error:",e.__class__)
```

```python
#Que5
""" ImportError- ImportError is raised when a module or member of module cannot be imported. 
    ModuleNotFoundError- This is the subclass of ImportError which is raised by import when a module could not be found. It is also raised 
    when None is found in sys.modules. """

# Example ImportError-
try:
    from time import myexception
except ImportError as e:
    print("Error:",e.__class__)
    
# Example ModuleNotFoundError-
try:
    import myexception
except ModuleNotFoundError as e:
    print("Error:",e.__class__)#Que5
```
