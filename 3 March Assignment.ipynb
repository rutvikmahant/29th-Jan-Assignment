```python
#Que1
'''Seaborn is a popular Python library used for data visualization. Here are five types of plots that can be created using Seaborn along with their uses:

scatterplot
histplot
relpolt
pairplot
catplot
Use:
1. scatterplot: This plot is used to show the relationship between two continuous variables. It can help in identifying any patterns or trends in the data

2. histplot: This plot is used to create a histogram of a variable or a dataset. A histogram is a type of plot that displays the distribution of continuous numerical data.

3. relpolt: This plot is used to create a relational plot, it is a flexible function that can craete different types of plots depending on the input data and type of plot to be created.

4. pairplot: This plot is used to create a pairwise plot for a dataset. It creates scatter plots between all pairs of variables in the dataset and histograms along the diagonal to show the distribution of each variable.

5. catplot: This is used to create several types of categorical plots such as bar plots, box plots, strip plots, violin plots, etc. It allows the user to specify the type of plot, the variables to be plotted, and the data to be used. can also be used to group data by one or more variables and to apply statistical functions to the groups.'''
```

```python
#Que2
import seaborn as sns

fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'])
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],data=fmri,hue='event')
```

```python
sns.lineplot(x=fmri['timepoint'],y=fmri['signal'],data=fmri,style='region')
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```
